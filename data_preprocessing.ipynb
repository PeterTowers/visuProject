```python
import pandas as pd
# import plotly.graph_objects as go
import plotly.express as px
```

```python
df = pd.read_csv('data/owid-covid-data.csv')
df.info()
```

```python
df_base = df[df['iso_code'].isin(["BRA", "MEX", "USA", "VNM", "KOR", "CHN", "GBR", "NZL", "AUS", "DEU", "FRA", "SWE"])]
df_base.info()
```

```python
cases = px.scatter(df_base, x='date', y='total_cases', color='location')
cases.show()
```

```python
new_cases_smooth = cases = px.scatter(df_base, x='date', y='new_cases_smoothed', color='location')
new_cases_smooth.show()
```

```python
deaths = px.scatter(df_base, x='date', y='total_deaths', color='location')
deaths.show()
```
